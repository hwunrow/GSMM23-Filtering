```python
def A_array(N):
    A=np.zeros((N,N))
    for i in range (N):
        for j in range (N):
            if (i+1)==j or (j+1)==i:
                A[i][j]=1
    return A

def A_circle(N):
    A=np.zeros((N,N))
    for i in range (N):
        for j in range (N):
            if (i+1)%N==j or (j+1)%N==i:
                A[i][j]=1
    return A

def A_full(N):
    A=np.zeros((N,N))
    for i in range (N):
        for j in range (N):
            if i!=j:
                A[i][j]=1
    return A

def A_dir(N):
    A=np.zeros((N,N))
    for i in range (N):
        for j in range (N):
            if i==(j-1)%N:
                A[i][j]=1
    return A
```
