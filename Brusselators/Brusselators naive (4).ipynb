```python
import time
import numpy as np
import matplotlib.pyplot as plt
import random
import time
from scipy.optimize import fsolve
import math
from sympy import Symbol, nsolve
from scipy.integrate import solve_ivp
from scipy.optimize import curve_fit
from numpy.random import rand
```

```python
#different choices of coupling

#array topology
def A_array(N):
    A=np.zeros((N,N))
    for i in range (N):
        for j in range (N):
            if (i+1)==j or (j+1)==i:
                A[i][j]=1
    return A

#circle topology
def A_circle(N):
    A=np.zeros((N,N))
    for i in range (N):
        for j in range (N):
            if (i+1)%N==j or (j+1)%N==i:
                A[i][j]=1
    return A

#all-to-all topology
def A_full(N):
    A=np.zeros((N,N))
    for i in range (N):
        for j in range (N):
            if i!=j:
                A[i][j]=1
    return A

#directed ring topology
def A_dir(N):
    A=np.zeros((N,N))
    for i in range (N):
        for j in range (N):
            if i==(j-1)%N:
                A[i][j]=1
    return A
```

```python

start=time.time()
print()



def run_function(par,zinit):
    #Define
    N=par[0]
    I=np.zeros(N)
    
    ti=par[1] #initial time
    tf=par[2] #final time
    t_inte=par[3] #number of steps

    c=par[4] #c parameter

    b=par[5] #b parameter

    d=par[6] #coupling

    A=A_full(N)

    #Build Laplacian
    L=np.zeros((N,N))
    for i in range (N):
        for j in range (N):
            if i==j:
                L[i][j]=sum(A[i][j] for j in range (N))
            else:
                L[i][j]=-A[i][j]

    # Define derivative function
    def f_der(t, z, b, c, d):
        dzdt=np.zeros(2*N)
        for i in range (N):
            dzdt[i] = 1-z[i]*(1+b-c*z[i]*z[i+N])

        for i in range (N):
            sum_coupling=0
            for j in range (N):
                sum_coupling+=L[i][j]*z[j+N]
            dzdt[i+N] = z[i]*(b-c*z[i]*z[i+N])-d*sum_coupling
        return dzdt


    # Define time spans and initial values

    tspan = np.linspace(ti, tf, t_inte)

    # Solve equations numerically
    sol = solve_ivp(lambda t, z: f_der(t, z, b, c, d), [tspan[0], tspan[-1]], zinit, t_eval=tspan, rtol = 1e-5)
    return sol

#par=(N,ti,tf,t_inte,c,b,d)

zinit=[1.53705918, 1.07878207, 0.47513891, 1.79234345, 0.93449475, 0.25445716,
     1.92848769, 1.07660849, 1.41892467, 0.46545651]#initial conditions

par=[5,0,200,100000,0.01,0.2,0.01]#parameters
solucao=run_function(par,zinit)

end=time.time()
print()
print("O tempo usado na solução foi: ")
print(end-start)
```
